```python
import pandas as pd
df_model = pd.read_csv("tcs_features_5d.csv")
print(df_model.head())
print(df_model.shape)
```

```python
FEATURES = [
    "ret_1d",
    "ret_5d",
    "ma_dist_10",
    "volatility_10d",
    "vol_chg_1d",
    "vol_ratio_10"
]

X = df_model[FEATURES]
y = df_model["target_5d"]
```

```python
split_idx = int(len(df_model) * 0.8)

X_train = X.iloc[:split_idx]
X_test  = X.iloc[split_idx:]

y_train = y.iloc[:split_idx]
y_test  = y.iloc[split_idx:]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

gbr.fit(X_train, y_train)
```

```python
gbr_preds = gbr.predict(X_test)

from sklearn.metrics import mean_absolute_error
import numpy as np

gbr_mae = mean_absolute_error(y_test, gbr_preds)
gbr_dir_acc = np.mean(
    np.sign(gbr_preds) == np.sign(y_test)
)

print("GB MAE:", gbr_mae)
print("GB Directional Accuracy:", gbr_dir_acc)
```
