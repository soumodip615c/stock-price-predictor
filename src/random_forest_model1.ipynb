```python
import pandas as pd
df_model = pd.read_csv("tcs_features_5d.csv")
print(df_model.head())
print(df_model.shape)
```

```python
print(df_model.columns)
```

```python
#['Date', 'Open', 'High', 'Low', 'Close', 'Volume',
 #'ret_1d', 'ret_5d', 'ma_10', 'ma_dist_10',
# 'volatility_10d', 'vol_chg_1d', 'vol_ma_10',
 #'vol_ratio_10', 'target_5d']
```

```python
FEATURES = [
    "ret_1d",
    "ret_5d",
    "ma_dist_10",
    "volatility_10d",
    "vol_chg_1d",
    "vol_ratio_10"
]

X = df_model[FEATURES]
y = df_model["target_5d"]
```

```python
split_idx = int(len(df_model) * 0.8)

X_train = X.iloc[:split_idx]
X_test  = X.iloc[split_idx:]

y_train = y.iloc[:split_idx]
y_test  = y.iloc[split_idx:]
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=6,
    min_samples_leaf=20,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
```

```python
preds = rf.predict(X_test)
```

```python
print(preds.shape)
```

```python
print(preds[:10])
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, preds)
print("MAE:", mae)
```

```python
import numpy as np

direction_accuracy = np.mean(
    np.sign(preds) == np.sign(y_test)
)

print("Directional Accuracy:", direction_accuracy)
```
